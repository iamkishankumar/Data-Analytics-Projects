# Scalling the data

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
df.columns
```

```python
X=df.drop(columns=["Total_turnover"])
y=df["Total_turnover"] 
```

```python
scaler=MinMaxScaler()
X_scaled=scaler.fit_transform(X)
```

# Data split

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, Y_train, Y_test= train_test_split(X_scaled,y, test_size=0.10,random_state=235)
```

```python
print(X_train.shape)
```

```python
print(Y_train.shape)
```

# Appling Linear Regression model to check the accuracy.

```python
from sklearn.linear_model import LinearRegression
```

```python
lr=LinearRegression()
lr.fit(X_train,Y_train)
```

```python
lr.intercept_
```

```python
lr.coef_
```

```python
Y_pred=lr.predict(X_test)
```

```python
Y_pred[0:5]
```

# Accuracy of the model

```python
from sklearn.metrics import mean_absolute_error, mean_absolute_percentage_error, mean_squared_error, r2_score
```

```python
r2_score(Y_test,Y_pred)
```

```python
mse=mean_squared_error(Y_test,Y_pred)
```

```python
np.sqrt(mse)
```

```python
mae= mean_absolute_error(Y_test,Y_pred)
mae
```

```python
sns.relplot(X_test[:,0],Y_test, color= "Red" )
sns.regplot(X_test[:,0],Y_pred, ci=False)
plt.show()
```

# Our model is given perfect prediction but still lets check model is there any change after applying regularisation.

```python
lasso=Lasso()
```

```python
lasso.fit(X_train,Y_train)
```
